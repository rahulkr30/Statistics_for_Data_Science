```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pareto

# Define parameters
alpha = 1
loc = 0
scale = 1


for i in range (1,5):

    dist = pareto(alpha, loc=loc, scale=scale)
    x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
    pdf = dist.pdf(x)
    alpha+=0.5
    scale+=2
    loc+=0.5
    plt.plot(x, pdf)

```
